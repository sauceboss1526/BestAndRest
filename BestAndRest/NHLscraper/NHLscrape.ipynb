Utilizing BeautifulSoup4


The website from which data is being acquired from (https://www.hockey-reference.com/draft/) is static, therefore, BeautifulSoup4 will be useful in acquiring respective data. However, the content from a website's html doc is needed for BeautifulSoup to scrape data from the website. 
"requests" is a package that requests for the html text behind the website. The html text can be passed into the BeautifulSoup scraper once the request is granted and urllib receives the data.

```python
import requests
#create urllib request
websiteurl = 'https://www.hockey-reference.com/draft/'
self = requests.get(websiteurl)

#pass the html in bs4 scraper
soup = BeautifulSoup(self.content, "html.parser")

#find data 
tablecols = soup.find('tr').text
year = soup.find('td', class_="left").text
rows = soup.find('table').text

#show data
print(rows)
```
