If we rerun the same code in the ipynb environmnet, the connection is successful and the data is returned as a dataframe with a dowloaded csv copy (check your folder)

```python
import pandas as pd
import numpy as np
import sodapy
from sodapy import Socrata 
def results(endpointdf, records, name):
        client = Socrata("data.cityofnewyork.us", None)
        data = client.get(endpointdf, limit=records)
        df = pd.DataFrame.from_records(data)
        df.to_csv(name)
        return df

results("jt7v-77mi", 2000, 'parkingtickets.csv')
```
